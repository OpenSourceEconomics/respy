# Replicating Keane and Wolpin (1997)

In the following, we will replicate the extended model from Keane and Wolpin (1997) and present occupational choice of agents over 40 periods.

```python
%matplotlib agg

import respy as rp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("../_static/respy.mplstyle")
```

```python
params, options = rp.get_example_model("kw_97_extended")
options["n_periods"] = 40
```

```python
state_space, df = rp.simulate(params, options)
```

```python
fig, ax = plt.subplots(figsize=(16, 10))

shares = df.groupby("Period").Choice.value_counts().unstack()[["home", "edu", "a", "b", "mil"]].div(1000)

shares.plot.bar(stacked=True, ax=ax, width=0.8)

ax.legend(
    labels=["Home", "School", "A", "B", "Military"],
    loc="upper center",
    bbox_to_anchor=(0.5, -0.1),
    ncol=5
)

ax.set_ylim(0, 1)

ax.set_xticks(range(0, 40, 5))
ax.set_xticklabels(range(0, 40, 5), rotation="horizontal")

ax.set_ylabel("Share of population");
```
